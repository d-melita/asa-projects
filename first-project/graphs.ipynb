```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import log
from scipy import stats
colors = ["#FF0B04", "#4374B3"]
sns.set_palette(sns.color_palette(colors))
```

```python
df = pd.read_csv("result-problem-1.csv")
df["nlogn"] = df["n"].apply(lambda x: x*np.log2(x)) # change "n" column to n*logn
df["n^2"] = df["n"].apply(lambda x: x*x) # change "n" column to n^2
df["time nlogn"] = df["time"].apply(lambda x: x*np.log2(x)) # change "time" column to time*logn
```

```python
ax1 = sns.set_style(style=None, rc=None)
plot1=sns.lineplot(x="nlogn", y="time", data=df, ax=ax1, color="red")
plot2=sns.scatterplot(x="nlogn", y="time", data=df, ax=ax1, color="blue")
plt.xlabel("N")
plt.ylabel("Time (s)")
```

```python
df2 = pd.read_csv("result-problem-2.csv")
df2["n*n"] = df2["n"].apply(lambda x: x*x)
df2["time n*n"] = df2["time"].apply(lambda x: x*x)
```

```python
ax1 = sns.set_style(style=None, rc=None)
sns.lineplot(x="n*n", y="time", data=df2, ax=ax1)
sns.scatterplot(x="n*n", y="time", data=df2, ax=ax1, color="blue")
plt.xlabel("N")
plt.ylabel("Time (s)")
#sns.lineplot(x="n", y="time", data=df2)
```

```python
plt.figure(figsize=(10,10))
# sns.regplot(x="n", y="time", data=df, color="blue")
slope, intercept, r_value, p_value, std_err = stats.linregress(df['nlogn'],df['time nlogn'])
# use line_kws to set line label for legend
ax = sns.regplot(x="nlogn", y="time nlogn", data=df, color='b', 
 line_kws={'label':"y={0:.1f}x+{1:.1f}".format(slope,intercept)})
 
print(r_value)

# plot legend
ax.legend()
plt.xlabel("N")
plt.ylabel("Time (s)")
#sns.lineplot(x="n", y="time", data=df2)
```
